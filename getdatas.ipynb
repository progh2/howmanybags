```python
import urllib.request
from bs4 import BeautifulSoup
import json

PRE_SCHOOL_CODE = "B10000"
START_NUM = 998
END_NUM = 1001

for i in range(START_NUM, END_NUM + 1 ):
    
    #find_word = '0501'
    find_word = '{:0=4}'.format(i)
    
    #url = 'http://m.schoolinfo.go.kr/disclosure/gongsi/infoDefault/B10000'
    url = urllib.request.Request("http://m.schoolinfo.go.kr/disclosure/gongsi01/" + PRE_SCHOOL_CODE + find_word + '/01')
    #url = urllib.request.Request("http://m.schoolinfo.go.kr/disclosure/gongsi/infoDefault/B10000" + find_word)
    # http://m.schoolinfo.go.kr/disclosure/gongsi/infoDefault/B10000
    data = urllib.request.urlopen(url).read().decode()
    bs = BeautifulSoup(data,'lxml') 
    #print(words)
    words = bs.find_all('h2',class_='title')
    
    if words:
        # 학교 이름 출력
        print( PRE_SCHOOL_CODE + find_word, words[0].text.strip())    
    
    data = bs.find_all('script')
    #m = re.search(r"JSON.parse\('([\w]+)'\);", str(data[10]))
    m = str(data[10]).split('JSON.parse(\'')[1]
    m = str(m).split('\');')[0]
    dict = json.loads(m)
    
    for data in dict:
        print("### " + data['displayName'] + " ###")      
        for items in data['dataList']:
            print("> " + items['key'] + ":" + items['count'])

    
```
